```python
import pandas as pd
import numpy as np
```

```python
from pandas_profiling import ProfileReport
```

```python
dataset = pd.read_csv("./Chinese_Company_dataset/cash_flow.csv")
```

```python
# Generate Report
profile = ProfileReport(dataset,
                        title="Chinese Companies Report",
                        correlations={
                                        "pearson": {"calculate": False},
                                        "spearman": {"calculate": False},
                                        "kendall": {"calculate": False},
                                        "phi_k": {"calculate": False},
    })
```

```python
profile.to_file("./Chinese_Company_dataset/company_report.html")
```
